```python
import pandas as pd
import re

def extract_text_in_parentheses(text):
    match = re.search(r'\((.*?)\)', text)
    if match:
        return match.group(1)
    else:
        return ''

input_file1 = 'ExperimentalHEB6.xlsx'
df1 = pd.read_excel(input_file1)
materials1 = df1.iloc[:, 0]
synthe = df1.iloc[:, 1]
# print(synthe)  

materials_list1 = []
for material in materials1:
    materials_i = extract_text_in_parentheses(material)
    materials_list1.append(materials_i)
# print(','.join(materials_list1))

input_file = 'HEB6.xlsx'
df = pd.read_excel(input_file)
df_copy = df.copy()
materials = df.iloc[:, 0]
materials_list = []
for material in materials:
    materials_i = extract_text_in_parentheses(material)
    materials_list.append(materials_i)

# input_text = input("Please input quinary HEB6 selected form Be, Mg, Ca, Sr, Ba, Sc, Y and lanthanide (excluding radioactive element), eg.BeMgCaSrLa,BeMgCaSrBa")
input_text = ','.join(materials_list1)
elements_list = input_text.split(",")

merged_df = pd.DataFrame()
for i in range(len(elements_list)):
    elements_list_i = re.findall('[A-Z][^A-Z]*', elements_list[i])
    a = set(elements_list_i)
    for j in range(len(materials_list)):
        materials_list_i = re.findall('[A-Z][^A-Z]*', materials_list[j])
        b = set(materials_list_i)
        if a == b:
            df_j = df.iloc[j, :]
            merged_df = pd.concat([merged_df, df_j], axis=1)
            print(j)
            df_modified = df_copy.drop(j, axis=0)
#             print(df.loc[j, 'Experiments'])
#             df.loc[j, 'Experiments'] = synthe[i]

# print(df_copy)
df_copy.to_csv('HEB6nwe.csv')
merged_df.T.to_excel('SearchResults.xlsx')

# merged_df.T
```
